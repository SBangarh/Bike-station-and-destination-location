Send a request to CityBikes for the city of your choice. 

```python
#imports
import requests
import pandas as pd
from pprint import pprint
import json
```

```python


url = "http://api.citybik.es/v2/networks/mobibikes"
citybikes = requests.get(url)
cb_data = citybikes.json()
#pprint(cb_data)
```

```python
#write to json file for later use
with open('citybikes_responses.json', 'w') as f:
    json.dump(cb_data, f)
```

```python
with open('citybikes_responses.json') as cb_json:
    cb_responses = json.load(cb_json)
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
stations = cb_responses['network']['stations']
keys = ['latitude', 'longitude', 'free_bikes']
cb_values = []
for station in stations:
    cb_values += [(key, station[key]) for key in keys]

pprint(cb_values)    

```

Put your parsed results into a DataFrame.

```python
cb_dataframe = pd.DataFrame(cb_values)
dict1 = {}
for row in cb_dataframe.itertuples():
    if row[1] not in dict1:
        dict1[row[1]] = []

for rows in cb_dataframe.itertuples():
    if rows[1] in dict1:
        dict1[rows[1]].append(rows[2])


cb_df_final = pd.DataFrame.from_dict(dict1)
cb_df_final
```

```python
# Created a list of list with latitude and longitude pairs for part 2 in the yelp_foursquare_EDA
cb_lat_long = []
for rows in cb_df_final.itertuples():
    lat_long_list = [rows[1], rows[2]]
    cb_lat_long.append(lat_long_list)
%store cb_lat_long
# 
```
